## New categorical columns

```python
reasons_column = pd.get_dummies(df['Reason for Absence'])
```

## Drop Reason for Absence

```python
df = df.drop(['Reason for Absence'], axis=1)
```

```python
reasons_column.head()
```

```python
reasons_column.loc[:, 1:14].max(axis=1)
```

## Group common reasons

```python
reason_type1 = reasons_column.loc[:, 1:14].max(axis=1)
reason_type2 = reasons_column.loc[:, 15:17].max(axis=1)
reason_type3 = reasons_column.loc[:, 18:21].max(axis=1)
reason_type4 = reasons_column.loc[:, 22:].max(axis=1)
```

```python
reason_type1
```

## Concatination

```python
df = pd.concat([df, reason_type1, reason_type2, reason_type3, reason_type4], axis=1)
df.head()
```

```python
df.columns.values
```

## Reorder columns

```python
column_names =['Date', 'Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index', 'Education',
       'Children', 'Pets', 'Absenteeism Time in Hours', 'Reason_1', 'Reason_2', 'Reason_3', 'Reason_4']

df.columns = column_names
```

```python
df.head()
```

```python
df.columns.values
```

```python
reordered_column = ['Reason_1',
       'Reason_2', 'Reason_3', 'Reason_4', 'Date', 'Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index', 'Education',
       'Children', 'Pets', 'Absenteeism Time in Hours'
       ]
```

```python
df = df[reordered_column]
```

```python
df.head()
```

## Create checkpoint

```python
df_reason_mod = df.copy()
```

```python
df_reason_mod.head()
```

## Date

```python
df_reason_mod['Date'] = pd.to_datetime(df_reason_mod['Date'], format='%d/%m/%Y')
```

```python
df_reason_mod['Date'].head()
```

## Month

```python
month_list = []
month_list
```

```python
for i in range(len(df_reason_mod)):
  month_list.append(df_reason_mod['Date'][i].month)
```

```python
month_list
```

```python
df_reason_mod['Month Value'] = month_list
```

```python
df_reason_mod.head()
```

## Extract the weekday

```python
def date_to_weekday(date):
  return date.weekday()
```

```python
df_reason_mod['Day of the week'] = df_reason_mod['Date'].apply(date_to_weekday)
```

```python
df_reason_mod.head()
```

```python
df_reason_mod = df_reason_mod.drop(['Date'], axis=1)
```

```python
df_reason_mod.columns.values
```

```python
new_columns = ['Reason_1', 'Reason_2', 'Reason_3', 'Reason_4',
       'Month Value', 'Day of the week','Transportation Expense', 
       'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index', 'Education',
       'Children', 'Pets', 'Absenteeism Time in Hours']

df_reason_mod = df_reason_mod[new_columns]
```

```python
df_reason_mod.head()
```

### Create Checkpoint


```python
df_reason_date_mod = df_reason_mod.copy()
```

## 'Children', 'Education' and 'Pets'

```python
df_reason_date_mod['Education'].unique()
```

```python
df_reason_date_mod['Education'] = df_reason_date_mod['Education'].map({1:0, 2:1, 3:1, 4:1})
```

```python
df_reason_date_mod['Education'].unique()
```

## Final Checkpoint
